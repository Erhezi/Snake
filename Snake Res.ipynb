```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import plotly.graph_objects as go
```

```python
with open('Result_12x12_4x4.json', 'r') as file:
    b1 = json.load(file)
with open('Result_14x14_4x4.json', 'r') as file:
    b2 = json.load(file)
with open('Result_16x16_4x4.json', 'r') as file:
    b3 = json.load(file)
with open('Result_18x18_4x4.json', 'r') as file:
    b4 = json.load(file)
with open('Result_20x20_4x4.json', 'r') as file:
    b5 = json.load(file)
```

```python
data1 = pd.DataFrame(b1, columns = ['apples', 'survival time', 'rewards'])
data2 = pd.DataFrame(b2, columns = ['apples', 'survival time', 'rewards'])
data3 = pd.DataFrame(b3, columns = ['apples', 'survival time', 'rewards'])
data4 = pd.DataFrame(b4, columns = ['apples', 'survival time', 'rewards'])
data5 = pd.DataFrame(b5, columns = ['apples', 'survival time', 'rewards'])
```

```python
# apples
fig = go.Figure()

fig.add_trace(go.Histogram(x=data1['apples'], name = label1))
fig.add_trace(go.Histogram(x=data2['apples'], name = label2))
fig.add_trace(go.Histogram(x=data3['apples'], name = label3))
fig.add_trace(go.Histogram(x=data4['apples'], name = label4))
fig.add_trace(go.Histogram(x=data5['apples'], name = label5))

fig.update_layout(barmode='overlay')
fig.update_traces(opacity=0.4)

#fig["data"][0].update(go.Histogram(marker={"color": 'rgb(10, 10, 80)'}))
fig.update_layout(title='Histogram of apples for 2500 games',
                   xaxis_title='iterations',
                   yaxis_title='apples eaten')

fig.show()




# survived iters
fig = go.Figure()

fig.add_trace(go.Histogram(x=data1['survival time'], name = label1))
fig.add_trace(go.Histogram(x=data2['survival time'], name = label2))
fig.add_trace(go.Histogram(x=data3['survival time'], name = label3))
fig.add_trace(go.Histogram(x=data4['survival time'], name = label4))
fig.add_trace(go.Histogram(x=data5['survival time'], name = label5))

fig.update_layout(barmode='overlay')
fig.update_traces(opacity=0.4)

fig.update_layout(title='Histogram of survival time for 2500 games',
                   xaxis_title='iterations',
                   yaxis_title='survival time')

fig.show()





# rewards
fig = go.Figure()

fig.add_trace(go.Histogram(x=data1['rewards'], name = label1))
fig.add_trace(go.Histogram(x=data2['rewards'], name = label2))
fig.add_trace(go.Histogram(x=data3['rewards'], name = label3))
fig.add_trace(go.Histogram(x=data4['rewards'], name = label4))
fig.add_trace(go.Histogram(x=data5['rewards'], name = label5))

fig.update_layout(barmode='overlay')
fig.update_traces(opacity=0.4)

fig.update_layout(title='Histogram of rewards for 2500 games',
                   xaxis_title='iterations',
                   yaxis_title='rewards')

fig.show()
```
